## **Concepts**

### **What is Open, High, Low, Close?**
[Learn more](https://www.quantstart.com/articles/understanding-equities-data/)



* The **open** price is the price at which the stock started trading each day.
* The **close** price is the last buy-sell order that was executed for the stock and agreed upon after a full day of trading. 
     * The closing price of a stock is considered to be the standard measure of value for a specific day.
     * Depending on the liquidity of the stock this may have occured in the final few seconds of trading or much earlier in the day.

* **High** refers to the highest price at which a stock was traded during the time period specified.
*  **Low** refers to lowest traded price.


The **Disparity** Index is a technical analysis indicator that measures the difference between a stock's **current price** and its **moving average**, expressed as a **percentage**. It helps traders identify potential overbought or oversold conditions. 

### **How is Volume Recorded?** 




**Volume** is a measure of the number of shares traded in a particular time period. 
* Each time shares are bought and sold the trade volume is recorded.
     * If the same shares are bought and sold multiple times, the volume of shares in each transaction is recorded. 
     * Volume doesn't reflect the number of available shares, it is a measure of share turnover, a count of all transactions that occur on every share. 
     * Every market exchange tracks their trade volume and provides estimates at an intraday frequency.
     * An accurate report of the total daily volume is provided the following day.

* Volume is often used as a technical analysis indicator.



* Trade volume is considered a measure of market strength. 
     * A **rising stock price** with **increasing volume** indicates a **healthy market**. 
     * The higher the volume of trades the more liquidity the stock has and the higher the volume of trades during a price change, the more significant that price change becomes.

### **Adjusted Close**

The **final column** in the daily data above is the **adjusted close** price. 
* This is the most useful column for analysing historic price changes. 
* It incorporates changes in the price of an equity resulting from dividend payouts, stock splits and from new shares being issued

### **The Effect of Stock Splits**

 A **stock split** is a multiplication or division of a company's share content, withoit a change in its market capitalisation.
 * **Forward Split**:  If a stock splits 2:1 then a company has doubled the number of shares by halving each share's value.
 * Reverse Split: In a **reverse split** existing shares owned by investors are replaced with a smaller number of shares with no effect on market capitalisation.
     * Example: For example a 1-for-2 reverse split replaces every two shares owned by an investor with a single share. If an investor owned 20 shares in a company before the split they would have 10 at the same value after the split.

* The **ratio** determines whether a split is forward or reverse. 
     * For example a 3-for-1 split is a forward split as the first number is larger than the second
* The **announcement date** is when the company publically announces the plans for the split. 
* The **record date** is the date on which shareholders need to own the stock to be eligible for the split however, this right is transferable for shares bought and sold before the effective date. 
* The **effective date** is the date on which the shares actually appear in an investors account.

### **Calculating Stock Split Adjustments**


All stock prices prior to the effective date of the split are multiplied by a factor which is calculated from the split ratio. In an N-for-M split this factor is calculated by $\frac{M}{N}$ \
The volume of stock is also adjusted in a similar manner but the calculation of the factor is reversed.
* For a 2-for-1 stock split the volume adjustment factor is $\frac{2}{1}$  = 1  . All stock volumes prior to the effective date would be multiplied by 2. These methods apply to both forward and reverse splits.


### **The Effect of Dividends**
* **Dividends** are a share of profits that a company pays to its shareholders.
* There are three main types of dividend:
     * A **regular dividend**: paid consistently over time, most often on a quarterly basis
     * A **special dividend**: a one-off payment to shareholders due to the accumulation of cash that has no immediate use
     * A **variable dividend**: these are sometimes paid off by companies that produce commoidities in addition to regular dividends. They occur at cosistent intervals but vary in amount

* The **dividend yield**: is the annual dividend per sahre divided by the share price.
     * The dividend yield is used by investors to quickly estimate how much they could earn by investing in a stock.
     * A stock with a share price of $10 that pays a dividend of $0.60 per share has a yield of 6%. 

* The **declaration date** is the date on which the dividend is publially declared by the company.
* This announcement includes a **record date** which is the date on which an investor must own shares on orders to recieve the dividend.
* The **ex-dividend date** is the date on which new shareholders are not elegible for the dividend
* The **payments date**is the day on which the income is recieved by the shareholders

### **Calculating Dividend Adjustments**


üìò **Dividend Adjustment Formula**

$\text{Adjustment Factor} = \frac{\text{Close}{\text{\_pre-div}} - \text{Dividend}}{\text{Close}{\text{\_pre-div}}}$

$\text{Adjusted Close} = \text{Close} \times \text{Adjustment Factor}$

Where:
*	Close_pre-div = Close price the day before the ex-dividend date
*	Dividend = Dividend paid on the ex-dividend date
*	This adjustment is applied to all previous prices before the ex-dividend date.

üßÆ Example Scenario: \

*	On 2025-05-03, a $2 dividend is issued.
*	Previous day‚Äôs close (2025-05-02) = $105
*	Adjustment Factor = (105 - 2) / 105 = 0.98095
*	Adjusted Close (2025-05-02 and earlier) = Close √ó 0.98095
*	You typically adjust all previous closes before the ex-dividend date.

### **Rights Offerings**


A **right offering** is an offer to buy new shares created by a company.
* It reduces the values to be represented to the public and make to current shareholders abd entitls them to purchase the new shares at this price.
* This additionally raise the company's capital.
* The **creation of new shares** **lowers the value of the existing shares** as it **increases supply** and **each share** now represents **a smaller amount** of the net profit.

**Subscription warrants** give shareholders the option (but not the obligation) to buy more shares at a set price within a limited time (usually 30 days). These rights can be sold before they expire and help offset the dilution caused by issuing new shares. The **rights issue ratio** (e.g., 1:2) means a shareholder can buy 1 new share for every 2 they already own.

(Prior) Adjusted Close = Close * Adjustment Factor
* The adjustment factor is calculated from the rights issue ratio.
* Example: 
     * A company announces a rights offering to shareholders with a rights issue ratio of 1:2. The adjustment factor is (1+2)/2 or 3/2 = 1.5. In this case all prior closing prices of the stock will be multiplied by 1.5.
* The **adjusted close** price alters the close price based on the effects of corporate actions. 
     * It makes a stock's performance easier to evaluate as well as allowing an investor to compare the performance of multiple assets. 
     * Since it does not suffer from price drops due to stock splits, it is more appropriate for backtesting any trading strategy rules.

### **Ticker Symbol Changes**



* Stock tickers are created when a company begins to trade publicly on an exchange. They identify the stock at the exchange level.
* The International Securities Identification Number (ISIN) which is a unique identifier for a security. 
* Stock tickers were first created by Standard & Poor (S&P) and are used across the world on every major exchange.
* The number of characters in a ticker is variable and dependant on the exchange. 
     * Stocks traded on the New York Stock Exchange (NYSE) have tickers made up of 3 characters while stocks traded on the NASDAQ have 4 characters.

* Ticker symbol changes can result from a company name change, a merger or acquistion, or a delisting or bankruptcy event. 
     * Ticker symbols will update automatically 
     * The change will become visible within two to three days and all historical data will be updated accordingly.

* Symbol changes due to a name change are regarded as price neutral.
     - There is no adjustment made to historic share price or volume.
     - Examples include when AOL became Time Warner and when Totalizer became International Business Machines (IBM).

* Mergers or acquisitions are often regarded as a positive change. The ticker of the acquired company generally becomes the ticker of the acquirer.
     - When **two companies merge**, the company being **acquired (buyed)** often disappears, and its stock is replaced by the acquirer‚Äôs stock or **cash**.
     - To compare stock prices **before and after** the merger, we need to **adjust the old prices** so they match the new structure.
     - This is done by using a ‚Äú**split ratio**‚Äù (like 0.8 or 1.2), which is just **a number** we multiply the old prices by to make them consistent with the new ones.
     - üí° Think of it like resizing old prices to fit the new company, just like with stock splits ‚Äî but **we don‚Äôt change the volume.**

* Another reason for changes to symbols is delisting or bankruptcy filing (in a negative light)
     - If the value of stock in a company **begins to fall** the company may eventually be delisted from the exchange.
     - Each exchange has its own rules regarding the delisting process but generally when a **stock** becomes **a penny stock** or trades **below $1 per share,** the **company** can **no longer be traded publicly**.
     - The ticker will be suffixed by a .pk, .ob or a .otcbb. 
          - The **pk** indicates that the stock is now available on the 'pink sheets' 
          - The **ob** or **otcbb** means that the stock is now traded on the over-the-counter bulletin board.
               - Example: The **NASDAQ** exchange has a special case of ticker symbol changes in listings up to Jan 2016. They **added a Q to the ticker for bankruptcy** and **an E for delinquent Securities and Exchange Commission (SEC)** filings. This practice has now been discontinued but may show in historic data.

* Add and Calculate '**Adjusted_Close_Approx**' column

```python
cursor = conn.cursor()

# Ensure correct DB context
cursor.execute("USE StockMarket;")

# Step 2: Check if the column exists
cursor.execute("""
SELECT 1 
FROM sys.columns 
WHERE Name = 'Adjusted_Close_Approx' 
  AND Object_ID = Object_ID('dbo.StockData');
""")
column_exists = cursor.fetchone()

# Step 3: Add column if it doesn't exist
if not column_exists:
    print("Column 'Adjusted_Close_Approx' does not exist. Adding and populating it.")
    cursor.execute("""
        ALTER TABLE dbo.StockData
        ADD Adjusted_Close_Approx FLOAT;
    """)
    conn.commit()
else:
    print("Column already exists. Proceeding to update it.")

# Step 4: Populate or update the column
cursor.execute("""
    UPDATE dbo.StockData
    SET Adjusted_Close_Approx = 
        CASE 
            WHEN [Stock Splits] > 0 THEN ([Close] - Dividends) / [Stock Splits]
            ELSE ([Close] - Dividends)
        END;
""")
conn.commit()
print("Column 'Adjusted_Close_Approx' populated or updated successfully.")
```

### Random Forest Regressor

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import timedelta
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import TimeSeriesSplit, RandomizedSearchCV
from scipy.stats import randint, uniform
import warnings

warnings.filterwarnings('ignore', category=FutureWarning)

# --- Feature Engineering Functions ---

def calculate_ema(df, column, span):
    return df[column].ewm(span=span, adjust=False).mean()

def calculate_rsi(df, column, window=14):
    delta = df[column].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / (loss + 1e-10) 
    rsi = 100 - (100 / (1 + rs))
    return rsi

def calculate_macd(df, column, fast_period=12, slow_period=26, signal_period=9):
    ema_fast = df[column].ewm(span=fast_period, adjust=False).mean()
    ema_slow = df[column].ewm(span=slow_period, adjust=False).mean()
    macd = ema_fast - ema_slow
    signal = macd.ewm(span=signal_period, adjust=False).mean()
    return macd, signal

def apply_feature_engineering(df):
    """Apply feature engineering to the DataFrame"""
    print("Applying feature engineering...")
    df['Date'] = pd.to_datetime(df['Date'])
    df = df.sort_values(by=['Brand_Name', 'Date'])

    df_list = []
    brand=df['Brand_Name'].unique()
   
    for brand_name in brand:
        df_brand_temp = df[df['Brand_Name'] == brand_name].copy()

        # Basic features
        df_brand_temp['HL_PCT'] = (df_brand_temp['High'] - df_brand_temp['Low']) / (df_brand_temp['Low'] + 1e-9)
        df_brand_temp['PCT_change'] = (df_brand_temp['Close'] - df_brand_temp['Open']) / (df_brand_temp['Open'] + 1e-9)
        df_brand_temp['Rolling_Mean_5'] = df_brand_temp['Close'].rolling(window=5).mean()
        df_brand_temp['Rolling_Mean_10'] = df_brand_temp['Close'].rolling(window=10).mean()
        df_brand_temp['Volume_Change'] = df_brand_temp['Volume'].pct_change().fillna(0)
        df_brand_temp['Volatility_Close_Std'] = df_brand_temp['Close'].rolling(window=5).std().fillna(0)

        # Technical indicators
        df_brand_temp['EMA_12'] = calculate_ema(df_brand_temp, 'Close', 12)
        df_brand_temp['EMA_26'] = calculate_ema(df_brand_temp, 'Close', 26)
        df_brand_temp['RSI'] = calculate_rsi(df_brand_temp, 'Close', 14)
        df_brand_temp['MACD'], df_brand_temp['MACD_Signal'] = calculate_macd(df_brand_temp, 'Close')

        df_list.append(df_brand_temp)

    df = pd.concat(df_list, ignore_index=True)
    df.fillna(0, inplace=True)
    print("Feature engineering complete.")
    return df

# --- Forecasting Function ---

def forecast_rf(model, last_known_data_row, scaler, features_list, target_col, num_days, X_train_unscaled):
    """Forecast future stock prices using a trained RandomForestRegressor"""
    forecasted_data = []
    current_data_row = last_known_data_row.copy()

    # Get the actual last date from the test set for proper forecast start
    # Assuming 'test_df' is accessible here or passed as an argument.
    # If not, you'd need to adjust how the initial forecast date is determined.
    # For this example, let's assume test_df is available in the scope or passed.
    
    # Ensure the initial forecast date is derived from the last actual date
    # and is a valid datetime object.
    
    # Get the last date from the last_known_data_row (which should be the last actual date)
    last_actual_date = current_data_row['Date'] 
    current_forecast_date = last_actual_date # Start forecasting from the day after the last actual date

    for i in range(num_days):
        # Increment the date for the next forecast day
        current_forecast_date += pd.Timedelta(days=1)
        
        # Skip weekends to forecast only on business days
        while current_forecast_date.weekday() >= 5: # Saturday (5) or Sunday (6)
            current_forecast_date += pd.Timedelta(days=1)

        # Prepare input features
        input_features_series = current_data_row[features_list].astype(float)
        input_features = input_features_series.values.reshape(1, -1)
        
        # Handle non-finite values before scaling
        if not np.all(np.isfinite(input_features)):
            for j, feature_val in enumerate(input_features[0]):
                if not np.isfinite(feature_val):
                    feature_name = features_list[j]
                    mean_of_feature_in_training = np.nanmean(X_train_unscaled[:, j])
                    input_features[0, j] = mean_of_feature_in_training if np.isfinite(mean_of_feature_in_training) else 0.0

        input_features_scaled = scaler.transform(input_features)

        # Predict the Close price
        predicted_close = model.predict(input_features_scaled)[0]

        # Store the forecast
        forecasted_data.append({
            'Date': current_forecast_date, # Use the single scalar date here
            'Predicted_Close': predicted_close,
            'Brand_Name': current_data_row['Brand_Name']
        })

        # Approximate features for next iteration
        new_row = current_data_row.copy()
        new_row['Date'] = current_forecast_date # Assign the single scalar date
        new_row[target_col] = predicted_close
        new_row['Open'] = predicted_close
        new_row['High'] = predicted_close * np.random.uniform(1.005, 1.015)
        new_row['Low'] = predicted_close * np.random.uniform(0.985, 0.995)
        new_row['Volume'] = current_data_row['Volume'] * np.random.normal(1.0, 0.05)
        new_row['Dividends'] = 0
        new_row['Stock Splits'] = 0

        # Recalculate derived features (using current_data_row for previous values)
        new_row['HL_PCT'] = (new_row['High'] - new_row['Low']) / (new_row['Low'] + 1e-9)
        new_row['PCT_change'] = (new_row['Close'] - new_row['Open']) / (new_row['Open'] + 1e-9)
        
        new_row['Rolling_Mean_5'] = (current_data_row['Rolling_Mean_5'] * 4 + new_row['Close']) / 5
        new_row['Rolling_Mean_10'] = (current_data_row['Rolling_Mean_10'] * 9 + new_row['Close']) / 10

        new_row['Volume_Change'] = (new_row['Volume'] - current_data_row['Volume']) / (current_data_row['Volume'] + 1e-9)
        
        new_row['Volatility_Close_Std'] = current_data_row['Volatility_Close_Std'] 
        
        ema12_multiplier = (2 / (12 + 1))
        ema26_multiplier = (2 / (26 + 1))
        new_row['EMA_12'] = (new_row['Close'] * ema12_multiplier) + (current_data_row['EMA_12'] * (1 - ema12_multiplier))
        new_row['EMA_26'] = (new_row['Close'] * ema26_multiplier) + (current_data_row['EMA_26'] * (1 - ema26_multiplier))

        new_row['RSI'] = current_data_row['RSI'] 
        new_row['MACD'] = current_data_row['MACD'] 
        new_row['MACD_Signal'] = current_data_row['MACD_Signal'] 

        # Final check for NaNs/Infinities
        for col_name in features_list:
            if not np.isfinite(new_row[col_name]):
                feature_index = features_list.index(col_name)
                replacement_value = np.nanmean(X_train_unscaled[:, feature_index])
                if not np.isfinite(replacement_value): replacement_value = 0.0
                new_row[col_name] = replacement_value

        current_data_row = new_row

    return pd.DataFrame(forecasted_data)

# --- Data Processing Functions ---

def process_brand_data(df, brand, features, target, future_days=90):
    """Process data for a single brand and return results"""
    print(f"\n--- Processing Brand: {brand.upper()} ---")
    
    brand_results = {}
    
    df_brand = df[df['Brand_Name'] == brand].copy().sort_values('Date')
    df_brand.replace([np.inf, -np.inf], np.nan, inplace=True)
    df_brand.dropna(inplace=True)

    if df_brand.shape[0] < len(features) + 20: 
        print(f"Insufficient data for {brand} after cleaning, skipping...")
        return None

    # Data Splitting
    train_date_end = '2022-12-31'
    val_date_end = '2023-12-31'

    train_df = df_brand[df_brand['Date'] <= train_date_end].copy()
    val_df = df_brand[(df_brand['Date'] > train_date_end) & (df_brand['Date'] <= val_date_end)].copy()
    test_df = df_brand[df_brand['Date'] > val_date_end].copy()

    if len(train_df) == 0 or len(val_df) == 0 or len(test_df) == 0:
        print(f"Not enough data in one of the splits for {brand}, skipping...")
        return None

    # Prepare data
    X_train = train_df[features].values
    y_train = train_df[target].values
    X_val = val_df[features].values
    y_val = val_df[target].values
    X_test = test_df[features].values
    y_test = test_df[target].values
    
    if not np.all(np.isfinite(X_train)):
        print(f"Non-finite values found in X_train for {brand}, skipping...")
        return None

    # Scaling
    scaler = MinMaxScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_val_scaled = scaler.transform(X_val)
    X_test_scaled = scaler.transform(X_test)

    # Model Training - Random Forest
    base_model = RandomForestRegressor(
        n_estimators=500,
        random_state=42,
        n_jobs=-1
    )

    # Random Forest hyperparameters
    param_distributions = {
        'max_depth': randint(3, 20),
        'min_samples_split': randint(2, 10),
        'min_samples_leaf': randint(1, 10),
        'max_features': ['sqrt', 'log2']
    }

    tscv = TimeSeriesSplit(n_splits=5)
    
    print(f"Starting RandomizedSearchCV for {brand} (Random Forest)...")
    rf_random = RandomizedSearchCV(
        estimator=base_model,
        param_distributions=param_distributions,
        n_iter=100,
        cv=tscv,
        verbose=1,
        random_state=42,
        n_jobs=-1,
        scoring='neg_mean_squared_error'
    )

    rf_random.fit(X_train_scaled, y_train)
    model = rf_random.best_estimator_

    # Make Predictions
    y_train_pred = model.predict(X_train_scaled)
    y_val_pred = model.predict(X_val_scaled)
    y_test_pred = model.predict(X_test_scaled)

    # Generate Forecasts
    last_actual_row = df_brand.iloc[-1].copy()
    rf_forecast_df = forecast_rf(model, last_actual_row, scaler, features, target, future_days, X_train)

    # Store results
    brand_results['train_df'] = train_df
    brand_results['val_df'] = val_df
    brand_results['test_df'] = test_df
    brand_results['y_train_pred'] = y_train_pred
    brand_results['y_val_pred'] = y_val_pred
    brand_results['y_test_pred'] = y_test_pred
    brand_results['forecast_df'] = rf_forecast_df
    brand_results['model'] = model
    brand_results['scaler'] = scaler
    
    # Calculate metrics
    brand_results['metrics'] = {
        'Brand_Name': brand,
        'Train RMSE': np.sqrt(mean_squared_error(y_train, y_train_pred)),
        'Train MAE': mean_absolute_error(y_train, y_train_pred),
        'Train R2': r2_score(y_train, y_train_pred),
        'Train MAPE (%)': np.mean(np.abs((y_train - y_train_pred) / (y_train + 1e-10))) * 100,
        'Validation RMSE': np.sqrt(mean_squared_error(y_val, y_val_pred)),
        'Validation MAE': mean_absolute_error(y_val, y_val_pred),
        'Validation R2': r2_score(y_val, y_val_pred),
        'Validation MAPE (%)': np.mean(np.abs((y_val - y_val_pred) / (y_val + 1e-10))) * 100,
        'Test RMSE': np.sqrt(mean_squared_error(y_test, y_test_pred)),
        'Test MAE': mean_absolute_error(y_test, y_test_pred),
        'Test R2': r2_score(y_test, y_test_pred),
        'Test MAPE (%)': np.mean(np.abs((y_test - y_test_pred) / (y_test + 1e-10))) * 100,
        'Model': model,
    }
    
    # Create prediction DataFrames
    test_predictions = pd.DataFrame({
        'Date': test_df['Date'],
        'Actual_Close': y_test,
        'Predicted_Close': y_test_pred,
        'Brand_Name': brand,
        'Type': 'Test'
    })
    
    forecast_predictions = rf_forecast_df.copy()
    forecast_predictions['Type'] = 'Future'
    
    brand_results['predictions'] = pd.concat([test_predictions, forecast_predictions], ignore_index=True)
    
    return brand_results

def process_all_brands(df, features, target, future_days=90):
    """Process data for all brands and return complete results"""
    brands = df['Brand_Name'].unique().tolist()
    
    all_results = {
        'summary': [],
        'predictions': [],
        'brand_data': {}
    }
    
    for brand in brands:
        brand_results = process_brand_data(df, brand, features, target, future_days)
        if brand_results is not None:
            all_results['summary'].append(brand_results['metrics'])
            all_results['predictions'].append(brand_results['predictions'])
            all_results['brand_data'][brand] = brand_results
    
    # Combine results
    all_results['summary_df'] = pd.DataFrame(all_results['summary'])
    all_results['predictions_df'] = pd.concat(all_results['predictions'], ignore_index=True)
    
    return all_results

```

```python
# --- Main Execution ---
#if __name__ == "__main__":
# --- Parameters ---
future_days = 90  # Number of days to forecast into the future
features = [
    'Open', 'High', 'Low', 'Volume', 'Dividends', 'Stock Splits',
    'HL_PCT', 'PCT_change', 'Rolling_Mean_5', 'Rolling_Mean_10',
    'Volume_Change', 'Volatility_Close_Std',
    'EMA_12', 'EMA_26', 'RSI', 'MACD', 'MACD_Signal'
]
target = 'Close'

# --- Data Preparation ---
# Assuming 'df' is loaded and contains the required columns
df = apply_feature_engineering(df)

# --- Process All Data ---
all_results = process_all_brands(df, features, target, future_days)

# --- Show Results ---
print("\n--- Summary of Model Performance Across Brands ---")
print(all_results['summary_df'])
print("\n--- All Predictions (Sample) ---")
print(all_results['predictions_df'].head())
print(all_results['predictions_df'].tail())

# Now you can plot whenever you want by calling:
# plot_all_results(all_results)
# Or plot individual brands with:
# plot_brand_results(all_results['brand_data'][brand_name])
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.backends.backend_pdf import PdfPages


# --- Plotting Functions ---

def plot_brand_results(brand_data, ax=None):
    """Plot results for a single brand"""
    if ax is None:
        fig, ax = plt.subplots(figsize=(10, 6))
    
    # Plot actual values
    ax.plot(brand_data['train_df']['Date'], brand_data['train_df']['Close'], 
            label='Train Actual', color='blue', alpha=0.7)
    ax.plot(brand_data['val_df']['Date'], brand_data['val_df']['Close'], 
            label='Validation Actual', color='green', alpha=0.7)
    ax.plot(brand_data['test_df']['Date'], brand_data['test_df']['Close'], 
            label='Test Actual', color='red', alpha=0.7)
    
    # Plot predicted values
    ax.plot(brand_data['val_df']['Date'], brand_data['y_val_pred'], 
            label='Validation Predicted', color='cyan', linestyle='--')
    ax.plot(brand_data['test_df']['Date'], brand_data['y_test_pred'], 
            label='Test Predicted', color='orange', linestyle='--')
    
    # Plot forecast
    ax.plot(brand_data['forecast_df']['Date'], brand_data['forecast_df']['Predicted_Close'], 
            label='RandomForestRegressor Forecast', color='purple', linestyle='-.')
    
    ax.set_title(f"{brand_data['metrics']['Brand_Name'].upper()} Price Prediction")
    ax.set_xlabel("Date")
    ax.set_ylabel("Close Price")
    ax.legend(fontsize='small')
    ax.grid(True)
    plt.setp(ax.get_xticklabels(), rotation=45)
    
    if ax is None:
        plt.tight_layout()
        plt.show()


def plot_all_results(all_results, n_cols=3, save_to_pdf=False, filename="brand_predictions.pdf"):
    """Plot results for all brands in a 3-column grid.
    
    Args:
        all_results (dict): Dictionary containing brand data.
        n_cols (int): Number of columns (default: 3).
        save_to_pdf (bool): If True, saves plots to a PDF file.
        filename (str): Output PDF filename (if save_to_pdf=True).
    """
    brands = list(all_results['brand_data'].keys())
    n_rows = int(np.ceil(len(brands) / n_cols))
    
    # Dynamically adjust figure height based on rows (to avoid squished plots)
    fig_height = 5 * n_rows  # 5 inches per row
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, fig_height))
    
    # Flatten axes for easy iteration
    axes = axes.ravel()
    
    # Plot each brand
    for i, brand in enumerate(brands):
        plot_brand_results(all_results['brand_data'][brand], ax=axes[i])
    
    # Hide unused subplots
    for j in range(i + 1, len(axes)):
        axes[j].axis('off')
    
    plt.subplots_adjust(
        hspace=0.5,  # Vertical spacing between subplots
        wspace=0.3   # Horizontal spacing
    )
    
    if save_to_pdf:
        with PdfPages(filename) as pdf:
            pdf.savefig(fig, bbox_inches='tight')
        plt.close(fig)
        print(f"Saved plots to {filename}")
    else:
        plt.tight_layout()
        plt.show()
```
